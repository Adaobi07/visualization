```python
import pandas as pd

#Load the uploaded CSV file
file_path = '../02_activities/assignments/Assignment4 data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
# Display the first few rows of the dataframe to understand its structure
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Convert the 'DATE' column to datetime format
data['DATE'] = pd.to_datetime(data['DATE'])

# Plotting the data
plt.figure(figsize=(12, 6))

# Group the data by age category and plot each group
for age_group, group_data in data.groupby('age_category'):
    plt.plot(group_data['DATE'], group_data['percent_positive_7d_avg'], label=age_group)

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Percent Positive (7-day Average)')
plt.title('Percent Positive Cases Over Time by Age Category')
plt.legend(title='Age Category')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
