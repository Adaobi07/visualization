```python
#import these libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50) #first value is our min, 2nd is our max, 3rd is our
#Creating a scattered diagram
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots (figsize=(5,3)) #This produces the size of the diagram
ax.plot(x,y) #to make it bar
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) #This produces the histogram. 
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time') #The title of the graph that will be produced
ax.set_ylabel('Total Growth') #The label on the y axis
ax.set_xlabel('Years since start') #The label on the x axis
```

```python
font1 ={'family':'san-serif','color':'blue','size':20}
font2 = {'family':'monospace', 'color': 'green', 'size' :14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# libraries
import numpy as np
from numpy import linspace
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# dataframe
df = pd.DataFrame({
'var1': np.random.normal(size=1000),
'var2': np.random.normal(loc=2, size=1000) * -1
})

# Fig size
plt.rcParams["figure.figsize"]=12,8

# plot density chart for var1
sns.kdeplot(data=df, x="var1",  fill=True, alpha=1)

# plot density chart for var2
kde = gaussian_kde(df.var2)
x_range = linspace(min(df.var2), max(df.var2), len(df.var2))

# multiply by -1 to reverse axis (mirror plot)
sns.lineplot(x=x_range*-1, y=kde(x_range) * -1, color='orange') 
plt.fill_between(x_range*-1, kde(x_range) * -1, color='orange')

# add axis names        
plt.xlabel("value of x")
plt.axhline(y=0, linestyle='-',linewidth=1, color='black')
 
```
